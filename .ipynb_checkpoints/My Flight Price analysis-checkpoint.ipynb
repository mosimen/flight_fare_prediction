```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost
import seaborn as sb
pd.set_option("display.max_columns",None)
```

```python
data=pd.read_excel("Data.xlsx")
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data["Total_Stops"].unique()
```

```python
data.loc[data["Total_Stops"].isnull()==True]
```

```python
data=data.dropna(axis=0, how='any')
```

```python
data.isnull().sum()
```

```python
data=data.drop(columns=["Route","Additional_Info"])
```

```python
data.head()
```

```python
data["Date_of_Journey"]=pd.to_datetime(data["Date_of_Journey"])

data["day_of_journey"]=data["Date_of_Journey"].dt.day
data["dayname_of_journey"]=data["Date_of_Journey"].dt.day_name()
data["month_of_journey"]=data["Date_of_Journey"].dt.month
data["monthname_of_journey"]=data["Date_of_Journey"].dt.month_name()
```

```python
data.head()
```

```python
data["Dep_Time"]=pd.to_datetime(data["Dep_Time"])

data["dep_hour"]=data["Dep_Time"].dt.hour
data["dep_minute"]=data["Dep_Time"].dt.minute
```

```python
data["Arrival_Time"]=pd.to_datetime(data["Arrival_Time"])

data["Arrival_hour"]=data["Arrival_Time"].dt.hour
data["Arrival_min"]=data["Arrival_Time"].dt.minute
```

```python
data.head()
```

```python
data["flight_hour"]=data["Duration"].apply(lambda x: x.split("h")[0] if "h" in x else 0)

data["flight_minutes"]=data["Duration"].apply(lambda x: 0 if "m" not in x else x.split("m")[0] if "h" not in x else x.split(" ")[1].split("m")[0])
```

```python
data["flight_hour"]=data["flight_hour"].astype('int')
data["flight_minutes"]=data["flight_minutes"].astype('int')
```

```python
data.head()
```

```python
data["Total_Stops"].unique()
```

```python
data["Total_Stops"]=data["Total_Stops"].replace({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4})
```

```python
data.head()
```

```python
data["dayname_of_journey"].unique()
```

```python
data["dayname_of_journey"]=data["dayname_of_journey"].replace({'Sunday':7, 'Saturday':6, 'Friday':5, 'Thursday':4, 'Monday':1, 'Tuesday':2,
       'Wednesday':3})
```

```python
data.head()
```

```python
data=data.drop(columns=["Date_of_Journey","Dep_Time","Arrival_Time","Duration","monthname_of_journey"])
```

```python
data.head()
```

```python
data=pd.get_dummies(data,columns=["Airline","Source","Destination"], drop_first=True)
```

```python
data.head()
```

```python
data.dtypes
```

```python
x=data.drop(columns="Price")

y=data["Price"]
```

```python
### Let us check the level of importance of the features or x columns

from sklearn.ensemble import ExtraTreesRegressor

model=ExtraTreesRegressor()
model.fit(x,y)
```

```python
model.feature_importances_

feature_imp=pd.Series(model.feature_importances_, index=x.columns)*100
feature_imp
```

```python
feature_imp.nlargest(20).plot(kind='barh')

plt.grid()
plt.show()
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV,cross_val_score
```

```python
trainx,testx,trainy,testy=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
trainx.shape,testx.shape,trainy.shape,testy.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
import xgboost
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import r2_score
```

```python
rfr=RandomForestRegressor()
xgr=xgboost.XGBRegressor()
```

```python
score_rf=cross_val_score(rfr,trainx,trainy,cv=10)
score_rf.mean()
```

```python
score_xg=cross_val_score(xgr,trainx,trainy, cv=10)
score_xg.mean()
```

```python
## Random Forest

rfr.fit(trainx,trainy)

yrf_pred=rfr.predict(testx)

r2_score(testy,yrf_pred)*100
```

```python
## Xg Boost

xgr.fit(trainx,trainy)

yxg_pred=xgr.predict(testx)

r2_score(testy,yxg_pred)*100
```

```python
rfh=RandomForestRegressor()
xgh=xgboost.XGBRegressor()
```

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]

# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations
bestrf = RandomizedSearchCV(estimator = rfh, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
bestrf.fit(trainx,trainy)
```

```python
yrfh_pred=bestrf.predict(testx)
r2_score(testy,yrfh_pred)*100
```

```python
import pickle

file=open('flightprice.pkl','wb')

pickle.dump(bestrf,file)
```
