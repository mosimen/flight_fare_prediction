```python
data=data.dropna(axis=0, how='any')
```

```python
data.isnull().sum()
```

```python
data=data.drop(columns=["Route","Additional_Info"])
```

```python
data.head()
```

```python
data["Date_of_Journey"]=pd.to_datetime(data["Date_of_Journey"])

data["day_of_journey"]=data["Date_of_Journey"].dt.day
data["dayname_of_journey"]=data["Date_of_Journey"].dt.day_name()
data["month_of_journey"]=data["Date_of_Journey"].dt.month
data["monthname_of_journey"]=data["Date_of_Journey"].dt.month_name()
```

```python
data.head()
```

```python
data["Dep_Time"]=pd.to_datetime(data["Dep_Time"])

data["dep_hour"]=data["Dep_Time"].dt.hour

data["dep_minute"]=data["Dep_Time"].dt.minute
```

```python
data["Arrival_Time"]=pd.to_datetime(data["Arrival_Time"])

data["Arrival_hour"]=data["Arrival_Time"].dt.hour
data["Arrival_min"]=data["Arrival_Time"].dt.minute
```

```python
data.head()
```

```python
data["flight_hour"]=data["Duration"].apply(lambda x: x.split("h")[0] if "h" in x else 0)

data["flight_minutes"]=data["Duration"].apply(lambda x: 0 if "m" not in x else x.split("m")[0] if "h" not in x else x.split(" ")[1].split("m")[0])
```

```python
data.head()
```

```python
data["Total_Stops"].unique()
```

```python
data["Total_Stops"]=data["Total_Stops"].replace({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4})
```

```python
data.head()
```

```python
data["dayname_of_journey"].unique()
```

```python
data["dayname_of_journey"]=data["dayname_of_journey"].replace({'Sunday':7, 'Saturday':6, 'Friday':5, 'Thursday':4, 'Monday':1, 'Tuesday':2,
       'Wednesday':3})
```

```python
data.head()
```

```python
data=data.drop(columns=["Date_of_Journey","Dep_Time","Arrival_Time","Duration","monthname_of_journey"])
```

```python
data.head()
```

```python
data=pd.get_dummies(data,columns=["Airline","Source","Destination"])
```

```python
data.head()
```

```python
data.shape
```

```python
x=data.drop(columns="Price")

y=data["Price"]
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
```

```python
trainx,testx,trainy,testy=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
trainx.shape,testx.shape,trainy.shape,testy.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
search={"n_estimators":[int(i) for i in np.linspace(50,2000,3)],
        "criterion":["mse", "mae"],
        "min_samples_split":[int(i) for i in np.linspace(2,10,3)],
        "min_samples_leaf":[int(i) for i in np.linspace(1,10,3)],
        "max_features":["auto", "sqrt", "log2"],
        "max_depth":[int(i) for i in np.linspace(10,200,3)],
       "bootstrap":[True,False]}
```

```python
bestrf=RandomizedSearchCV(estimator=rf,param_distributions=search,n_iter=10,cv=5,n_jobs=-1,verbose=2,random_state=42)
```

```python
bestrf.fit(trainx,trainy)
```
